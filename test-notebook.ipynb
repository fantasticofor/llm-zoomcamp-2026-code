```python
from dotenv import load_dotenv
load_dotenv()
from openai import OpenAI
openai_client = OpenAI()
def llm(prompt):
    response = openai_client.responses.create(
        model="gpt-5.4-mini",
        input=prompt
    )
    return response.output_text
```

```python
llm("Hey, what's up?")
```

```python
question = "I just discovered the course. Can I join now?"
answer = llm(question)
print(answer)
```
